## ROCK-PAPER-SCISSORS GAME

## README
#### INTRODUCTION:
Welcome to the Rock-Paper-Scissors game! This simple console-based game allows you to play the classic Rock-Paper-Scissors against the computer. The game continues for 15 rounds.

#### HOW TO PLAY:
Run the Python script.
Enter your name when prompted.
Choose your move by entering 'R' for rock, 'P' for paper, or 'S' for scissors.
The computer will randomly select its move.
The winner of each round will be displayed, and the game will continue for 15 rounds.
After the game concludes, you will see the overall statistics including the number of rounds played, wins, losses, and ties.

#### INPUT VALIDATION:
Ensure that you enter a valid move by choosing 'R', 'P', or 'S'. You forfeit a round if your move is invalid.

#### CODE STRUCTURE:
The game script is structured as follows:

Player input and game loop. 
Move validation and computer move generation. 
Determining the winner of each round. 
Displaying game statistics.

```python
import random

print("Welcome Player!")
player_name = str(input("Please tell us your name: "))
print ( )
    
player_wins = 0
computer_wins = 0
ties = 0
total_rounds = 0

while total_rounds < 15:
    
    user_choice = input("Enter 'R' for rock, 'P' for paper or 'S' for scissors: ").upper()

    computer_choice = random.choice(["R", "P", "S"])
    choices = ["R", "P", "S"]


    if user_choice == 'R':
        print (player_name + " played: ROCK")
    elif user_choice == 'P':
        print (player_name + " played: PAPER")
    elif user_choice == 'S':
        print (player_name + " played: SCISSORS")
    else:
        print("Invalid choice. You forfeit this round.")

    if computer_choice == 'R'and (user_choice == 'R'or user_choice == 'P' or user_choice == 'S'):
        print ("Computer played: ROCK")
    elif computer_choice == 'P'and (user_choice == 'R'or user_choice == 'P' or user_choice == 'S'):
        print ("Computer played: PAPER")
    elif computer_choice == 'S'and (user_choice == 'R'or user_choice == 'P' or user_choice == 'S'):
        print ("Computer played: SCISSORS")
    else:
        print ('Try again.')

    if user_choice == computer_choice:
        print( "It's a tie")
        print ( )
        ties +=1
        
    elif (user_choice == "R" and computer_choice == "S") or (user_choice == "P" and computer_choice == "R") or (user_choice == "S" and computer_choice == "P"):
        print( "You won")
        print ( )
        player_wins += 1
        
    elif (user_choice == "S" and computer_choice == "R") or (user_choice == "R" and computer_choice == "P") or (user_choice == "P" and computer_choice == "S"):
        print( "You lost")
        print ( )
        computer_wins += 1
    else:
        print ( )
        
    total_rounds += 1
    
    total = player_wins + computer_wins + ties
    
print ( )       
print ('GAME OVER!')
print ( )
print ('GAME STATISTICS:')
print ( )
print ('Rounds played:' + str(total))
print ('Wins:' + str(player_wins))
print ('Losses:' + str(computer_wins))
print ('Ties:' + str(ties))
```
